```python
#top 9 dataframe
top9 = pd.DataFrame(master_frame[master_frame['funding_round_type'] == 'private_equity'].groupby('country_code').
             raised_amount_usd.sum()).sort_values(by = 'raised_amount_usd', ascending = False)[:9]
```

```python
top9
```

```python
top9_countries = list(top9.index)
```

```python
top9_countries
```

```python

#top 9 countries and funding type private equity
top9_df = master_frame[(master_frame['country_code'].isin(top9_countries)) & (master_frame['funding_round_type'] == 'private_equity')]
top9_df
top9_df.shape
```

```python
#splitting the top 9 countries into the primary sector
top9_df['primary_sector'] = top9_df['category_list'].str.split('|', n = 2, expand = True)[[0]]
top9_df
```

```python
#import mapping csv
mapping_csv = pd.read_csv("C:\\Users\\Bindiya\\Desktop\\InvestmentAssignment\\mapping.csv")
```

```python
mapping_csv
```

```python
mapping_csv.drop(0, inplace = True)#dropping first row
```

```python

#melting mapping csv
mapping_csv = pd.melt(mapping_csv, id_vars=['category_list'])
```

```python
mapping_csv
```

```python
mapping_csv = mapping_csv[mapping_csv['value'] ==1] #retaining relevant rows
```

```python
mapping_csv.drop('value', axis = 1, inplace = True) #removing additional column
```

```python
#renaming columns
mapping_csv.rename(columns={'category_list':'primary_sector', 'variable':'main_sector'}, inplace = True)
```

```python
mapping_csv
```

```python
# merging maping and top9 df using left join to retain all records of top9_df
top9_df = pd.merge(top9_df, mapping_csv, how = 'left', on = 'primary_sector')
top9_df
```

```python
top9_df.columns
```

```python
top9_df[top9_df['main_sector_y'].isnull()]
```

```python
top9_df.loc[top9_df['primary_sector'].str.contains('Analytics', na = False),['main_sector_y']] = 'Social, Finance, Analytics, Advertising'
```

```python
top9_df.loc[top9_df['primary_sector'].str.contains('Financ', na = False),['main_sector_y']] = 'Social, Finance, Analytics, Advertising'
```

```python
#imputing null values
top9_df.loc[top9_df['main_sector_y'].isnull(), ['main_sector_y']] = 'Others'
```

```python
len(top9_df[top9_df['main_sector_y'].isnull()])
```

```python

# USA dataframe
USA_df = top9_df[(top9_df['country_code'] == 'USA') & ((top9_df['raised_amount_usd'] >= 5000000) & (top9_df['raised_amount_usd'] <= 15000000))]
```

```python
#USA sector_analysis
USA_sector_analysis = USA_df.groupby('main_sector_y')['raised_amount_usd'].agg(('sum', 'count'))
```

```python
#renamingcolumns
USA_sector_analysis.rename(columns={'sum':'Sum of investment', 'count': 'Count of Investment'}, inplace = True)
```

```python
# count and investment amount
USA_df = pd.merge(USA_df, USA_sector_analysis, on = 'main_sector_y')
```

```python
USA_df
```

```python
#Creating d2 India
India_df = top9_df[(top9_df['country_code'] == 'IND') & ((top9_df['raised_amount_usd'] >= 5000000) & (top9_df['raised_amount_usd'] <= 15000000))]
India_df
```

```python
# Sector analysis for India
India_sector_analysis = India_df.groupby('main_sector_y')['raised_amount_usd'].agg(('sum', 'count'))
```

```python

#renaming columns
India_sector_analysis.rename(columns={'sum':'Sum of investment', 'count': 'Count of Investment'}, inplace = True)
```

```python
#merging count and investment amount
India_df = pd.merge(India_df, India_sector_analysis, on = 'main_sector_y')
India_df
```

```python
# GBR
GBR_df = top9_df[(top9_df['country_code'] == 'GBR') & ((top9_df['raised_amount_usd'] >= 5000000) & (top9_df['raised_amount_usd'] <= 15000000))]
```

```python
GBR_df
```

```python
# creating sector analysis for d3
GBR_sector_analysis = GBR_df.groupby('main_sector_y')['raised_amount_usd'].agg(('sum', 'count'))
```

```python

#renaming columns
GBR_sector_analysis.rename(columns={'sum':'Sum of investment', 'count': 'Count of Investment'}, inplace = True)
```

```python

#merging count and investment amount
GBR_df = pd.merge(GBR_df, GBR_sector_analysis, on = 'main_sector_y')
```

```python
GBR_df
```

```python

# concatinating USA, INDIA AND GBR
top3 = pd.concat([USA_df, India_df, GBR_df], axis = 0, ignore_index = True, sort = False)
```
